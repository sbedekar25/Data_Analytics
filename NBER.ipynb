## Downloading all the data

```python
random.seed(16470221)
random.shuffle(ppl)
```

```python
def get_person(pid):
    html = requests.get('https://www.nber.org/people/' + pid).text
    
    name = re.search(r'<span itemprop="name">(.+?)</span>', html).group(1)
    
    m = re.search(r'NBER Affiliation: (.+?)<br />', html)
    nber_affil = m.group(1).strip() if m else ''
    
    m = re.search(r'Institutional Affiliations?: (.+?)<br />', html)
    inst_affil = m.group(1).strip() if m else ''

    m = re.search(r'addr=(.+?)"', html)
    email = m.group(1) if m else ''
    if email != '':
        email = ''.join(map(lambda x: sdr.get(x,x), email))
    
    m = re.search(r'WWW: <a href="(.+?)"', html)
    homepage = m.group(1) if m else ''

    # programs
    m = re.search(r'NBER Program Affiliations:(.+?)<br>', html, re.DOTALL)
    if m:
        prog = m.group(1)
        progs = re.findall(r'/programs/(\w+)', prog)
    else:
        progs = []
    
    # papers/coauthors
    papers = []
    tbl = re.search(r'<table>.+?</table>', html, re.DOTALL).group(0)
    rows = re.findall(r'<tr>(.+?)</tr>', html, re.DOTALL)
    for row in rows:
        try:
            m = re.search(r'/(?:papers|chapters)/(\w\d+)">(.+?)</a>', row)
            paper_id = m.group(1)
            paper_title = m.group(2)
            # remove white space
            paper_title = ' '.join(paper_title.split())
            
            paper_date = re.search(r'<nobr>(\w+ \d+)</nobr>', row).group(1)
            
            coauthors = re.findall(r'/people/(.+?)"', row)
            
            papers.append((paper_id, paper_date, paper_title, coauthors))
        except:
            pass

    return ((pid, name, nber_affil, inst_affil, email, homepage, progs), papers)
```

```python
people = []
papers = {}

# for pid in random.sample(ppl, 3):
for i,pid in enumerate(ppl):
    if i%10==0:
        print('{}  {}: {}'.format(time.strftime("%H:%M:%S"), i, pid))
    result = get_person(pid)
    people.append(result[0])
    papers.update({pid : result[1]})
    time.sleep(1.5 + random.random())
```
